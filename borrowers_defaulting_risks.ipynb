### Step 2. Data preprocessing

#### Missing Values

```python
# Change the negatives in 'days_employed' to positives 
credit_worthiness['days_employed'] = credit_worthiness['days_employed'].abs()

# Replace the NaNs with days_employed_median since the mean shows significant outlier(s).
days_employed_median = credit_worthiness['days_employed'].median()
credit_worthiness['days_employed'] = credit_worthiness['days_employed'].fillna(days_employed_median)

# Replace the 0s with dob_median since the mean shows significant outlier(s).
credit_worthiness.loc[credit_worthiness['dob_years'] == 0, 'dob_years'] = credit_worthiness['dob_years'].median()

# Replacethe NaNs in total_income column with the mean grouped by 'income_type' since there is a higher chance that 
# people in the same category of income type will the same total income.
total_income_mean = credit_worthiness.groupby('income_type')['total_income'].transform('mean')
credit_worthiness['total_income'] = credit_worthiness['total_income'].fillna(total_income_mean )
```

```python
#change the values of '-1' and '20' in 'children' column to 1 and 2 respectively
credit_worthiness.loc[credit_worthiness['children'] == -1, 'children'] = 1
credit_worthiness.loc[credit_worthiness['children'] == 20, 'children'] = 2
credit_worthiness['children'].value_counts()
```

```python
display(credit_worthiness.info())
```

### Data type replacement

```python
credit_worthiness['total_income'] = credit_worthiness['total_income'].astype(int)
credit_worthiness['days_employed'] = credit_worthiness['days_employed'].astype(int)
credit_worthiness['dob_years'] = credit_worthiness['dob_years'].astype(int)
display(credit_worthiness.head())
display(credit_worthiness.info())
```

### Processing duplicates

```python
# Drop the duplicates
credit_worthiness = credit_worthiness.drop_duplicates(inplace=False)
credit_worthiness.info()
```

### Categorizing Data

```python
from nltk.stem import SnowballStemmer 
english_stemmer = SnowballStemmer('english')


def purpose_category_eng(queries):
    '''
    This function returns a word according to the stem word in the given word or phrase
    '''
    
    for word in queries.split(' '):
        stemmed_word = english_stemmer.stem(word)
        if stemmed_word == 'hous':
            return 'housing'
        if stemmed_word == 'car':
            return 'car purchase'
        if stemmed_word == 'educ' or stemmed_word == 'univers':
            return 'education'
        if stemmed_word == 'wed':
            return 'wedding'
        if stemmed_word == 'estat':
            return 'real estate'
        if stemmed_word == 'properti':
            return 'property'
    return 'unknown'


credit_worthiness['purpose_category'] = credit_worthiness['purpose'].apply(purpose_category_eng)
```

- **Is there a relation between having kids and repaying a loan on time?**

```python
childdebt = credit_worthiness.groupby('children').agg({'debt': ['count', 'sum']}).reset_index()

childdebt.columns = ('Number of Children', 'num_of_clients', 'num_of_debtors')
childdebt['percentage'] = ((childdebt['num_of_debtors'] / childdebt['num_of_clients']).round(4)) * 100
display(childdebt)

# Graphical Representation
def graphs(df, x, y):
    plt.figure(figsize=(8,6))
    sns.barplot(data=df, x=x, y=y).set_title('{} and Proportion of Debtors'.format(x))
    plt.xticks(rotation=85)
    plt.show()

graphs(childdebt, 'Number of Children', 'percentage')
```

### Conclusion
From the data, I can say yes there is a relationship between having kids and repaying a loan on time. This is because the rate of debt owed by those with one to four kids are higher than the rate of debt owed by those without kids. Those with five kids have no debt owed. Since those with five kids are 0.064% of the population in question, it can be ignored

- **Is there a relation between marital status and repaying a loan on time?**

```python
familydebt = (
            credit_worthiness
            .groupby('family_status')
            .agg({'debt': ['count', 'sum']})
            .reset_index()
            )
familydebt.columns = ('Family Status', 'num_of_clients', 'num_of_debtors')
familydebt['percentage'] = ((familydebt['num_of_debtors'] / familydebt['num_of_clients']) * 100).round(2)
display(familydebt)

# Graphical Representation
graphs(familydebt, 'Family Status', 'percentage')
```

### Conclusion
Yes, there is a relationship between marital status and loan repayment. While the widow/widower group has the lowest repayment rate, unmarried group has the highest repayment rate. I can conclude that some groups are more dependable when it comes to loan repayment than others. Their is a relationship between the two variables.

- **Is there a relation between income level and repaying a loan on time?**

```python
# Create ranges of income from 'total_income' column
credit_worthiness['quantiles_5'] = pd.qcut(x=credit_worthiness['total_income'], q=5)

incomedebt = (
            credit_worthiness
            .groupby('quantiles_5')
            .agg({'debt': ['count', 'sum']})
            .reset_index()
            )
incomedebt.columns = ('Income Range', 'num_of_clients', 'num_of_debtors')
incomedebt['percentage'] = ((incomedebt['num_of_debtors'] / familydebt['num_of_clients']) * 100).round(2)
display(incomedebt)

# Graphical Representation
graphs(incomedebt, 'Income Range', 'percentage')
```

**Conclusion**

Yes, there is a relationship between income level and repaying a loan on time. The group of customers that earn between 3305.999 and 15765.4 and between 21670.8 and 26022.2 have a greater tendency of repaying loans on time than the group of customers that earn between 15765.4 and 21670.8 and between 34339.4 and 362496.0.

- How do different loan purposes affect on-time repayment of the loan?

```python
purposedebt = (
            credit_worthiness
            .groupby('purpose_category')
            .agg({'debt': ['count', 'sum']})
            .reset_index()
            )
purposedebt.columns = ('Purpose', 'num_of_clients', 'num_of_debtors')
purposedebt['percentage'] = ((purposedebt['num_of_debtors'] / purposedebt['num_of_clients']) * 100).round(2)
display(purposedebt)

# Graphical Representation
graphs(purposedebt, 'Purpose', 'percentage')
```

### Conclusion
People who want loan for car purchase and education have higher debt rate than people who want loan for housing. In other words, people who want loan for housing have a better chance at repaying their loan on time.
